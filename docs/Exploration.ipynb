````python
import os
import pandas as pd
import matplotlib.pyplot as plt
````

````python
ship_stats_file=os.path.join('../data/ShipStats.csv')
ship_stats_df=pd.read_csv(ship_stats_file, sep=';')
# Drop unnecessary columns
ship_stats_df.drop(ship_stats_df.columns[[1, 2, 8, 9, 10]], axis=1, inplace=True)
# Drop non-warships
ship_stats_df=ship_stats_df[ship_stats_df['Dmg per shot'].notnull()]
# Cast columns correctly
ship_stats_df['Dmg per shot']=ship_stats_df['Dmg per shot'].apply('int64')
ship_stats_df['Attack Range']=ship_stats_df['Attack Range'].apply('int64')
# Reset index after removals
ship_stats_df=ship_stats_df.reset_index(drop=True)

ship_stats_df.head()
````

````python
ship_stats_df.sort_values('Hitpoints').plot(x='Type', y='Hitpoints', kind='bar')
ax=ship_stats_df['Dmg per second'].plot(secondary_y=True, color='red', rot=90)
ax.set_ylabel('Dmg per second')
plt.show()
````

````python
# Pandas Dataframe.rank() method returns a rank of every respective index of a series passed. The rank is returned on the basis of position after sorting.
ship_stats_df['Hitpoints Rank']=ship_stats_df['Hitpoints'].rank().astype(int)
ship_stats_df['Dmg per second Rank']=ship_stats_df['Dmg per second'].rank().astype(int)
ship_stats_df.style\
    .bar(subset=['Hitpoints Rank', 'Dmg per second Rank'])\
    .format(precision=2)
````

````python
ship_stats_df['HPxDMG/sec Rank']=((ship_stats_df['Hitpoints']*ship_stats_df['Dmg per second'])/2000).astype(int)
ship_stats_df['HPxDMG/secxRange Rank']=((ship_stats_df['Hitpoints']*ship_stats_df['Dmg per second']*ship_stats_df['Attack Range'])/3000).astype(int)
ship_stats_df.style\
    .background_gradient(subset=['HPxDMG/sec Rank', 'HPxDMG/secxRange Rank'], cmap ='Greens')\
    .format(precision=2)
````
